```python
# Stack array vertically
import numpy as np
data1 = np.arange(10).reshape(2, -1)
print(data1)

data2 = np.repeat(7, 10).reshape(2, -1)

print(data2)

# Way 1: stack data1 on top data2
out1 = np.concatenate([data1, data2], axis=0)
print(out1)

# Way 2: verticall stack
out2 = np.vstack([data1, data2])
print(out2)

# Way 3: np.r_
out3 = np.r_[data1, data2]
```

```python
# Stack array horizontally

# Way 1: np.concaternate
out4 = np.concatenate([data1, data2], axis=1)
print(out4)

# Way 2: np.hstack
out5 = np.hstack([data1, data2])
print(out5)

# Way 3: np.c_
out6 = np.c_[data1, data2]
print(out6)
```

```python
# Apply user-defined function to np.vectorize
import numpy as np

data1 = np.array([5, 3, 8, 2, 7])
data2 = np.array([2, 7, 3, 1, 8])

def get_max(a, b):
    if a >= b:
        return a
    return b

pair_max = np.vectorize(get_max, otypes=[int])
out = pair_max(data1, data2)
print(out)
```

```python
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt

img = mpimg.imread('dog.jpeg')


height, width, channel = img.shape
grey_img = np.zeros((height, width))

for h in range(height):
    for w in range(width):
        pixel = img[h][w]
        grey = np.mean(pixel)
        grey_img[h][w] = grey

f, axarr = plt.subplots(2, 1)
axarr[0].imshow(img)
axarr[1].imshow(grey_img, cmap='gray')
print(grey_img[0, 0])
```

```python
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt

# Gray scale using luminosity

img = mpimg.imread('dog.jpeg')

height, width, channel = img.shape
grey_img = np.zeros((height, width))

for h in range(height):
    for w in range(width):
        pixel = img[h][w]
        r, g, b = pixel
        grey = 0.21 * r + 0.72 * g + 0.07 * b
        grey_img[h][w] = grey

f, axarr = plt.subplots(2, 1)
axarr[0].imshow(img)
axarr[1].imshow(grey_img, cmap='gray')
print(grey_img[0, 0])
```

```python
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt

# Gray scale using lightness

img = mpimg.imread('dog.jpeg')

height, width, channel = img.shape
grey_img = np.zeros((height, width))

for h in range(height):
    for w in range(width):
        pixel = img[h][w]
        r, g, b = pixel
        grey = (max(r, g, b) + min(r, g, b)) / 2
        grey_img[h][w] = grey

f, axarr = plt.subplots(2, 1)
axarr[0].imshow(img)
axarr[1].imshow(grey_img, cmap='gray')
print(grey_img[0, 0])
```

```python
def categorize(a, value):
    if a > value:
        return 'Good'
    elif a < value:
        return 'Bad'
    return 'Average'


def question20():
    # Get sales column
    sales_column = df['Sales']

    # mean_sales
    mean_sales = df['Sales'].mean()
    print('mean sales', mean_sales)

    pair_cat = np.vectorize(categorize, otypes=[str])

    
    print('sales column', sales_column)

    scores = pair_cat(sales_column, mean_sales)
    print(scores[7:10])
    print(sales_column[7:10])
    



question20()
```

```python
def question21():
    sales_column = df['Sales']
    mean_sales = sales_column.mean()
    
    diffs = abs(sales_column - mean_sales)
    least_diff = np.min(diffs)

    value_to_compare = mean_sales + least_diff
    
    pair_cat = np.vectorize(categorize, otypes=[str])
    
    scores = pair_cat(sales_column, value_to_compare)

    print(scores[7:10])

question21()
```
