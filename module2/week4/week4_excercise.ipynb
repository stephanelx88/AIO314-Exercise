```python
# Question 1
import numpy as np

def compute_mean(x):
    total = np.sum(x)
    length = len(x)

    return total / length


x = [2, 0, 2, 2, 7, 4, -2, 5, -1, -1]

print('Mean:', compute_mean(x))
```

```python
# Question 2:

def compute_median(x):

    # Sort x ascending
    sorted_x = sorted(x)

    # Median
    median = None
    if len(x) % 2 != 0:
        mid = int(len(x) / 2)
        
        median = sorted_x[mid]
    else:
        first_index = int(len(x) / 2 - 1)
        second_index = first_index + 1
        
        median = (sorted_x[first_index] + sorted_x[second_index]) / 2

    return median

x = [1, 5, 4, 4, 9, 13]

print('Median:', compute_median(x))
```

```python
# Question 3:
import numpy as np


def compute_std(x):
    
    # Compute mean
    mean = compute_mean(x)

    # Variance
    variance = 0

    for num in x:
        variance += (num - mean)**2

    variance = variance / len(x)

    std = np.sqrt(variance)

    return std


x = [ 171 , 176 , 155 , 167 , 169 , 182]

print(compute_std(x))
```

```python
# Question 4:
import numpy as np

def compute_correlation_coefficient(x, y):
    n = len(x)
    numerator = 0
    denominator = 0
   

    numerator = n * (np.sum(x*y)) -  (np.sum(x) * np.sum(y))
    denominator = np.sqrt((n*np.sum(x**2) - np.sum(x)**2) * (n*np.sum(y**2) - np.sum(y)**2))

    return np.round(numerator / denominator, 2)


# Tests
x = np.asarray([-2, -5, -11, 6, 4, 15, 9])
y = np.asarray([4 , 25, 121 , 36, 16, 225 , 81])

print('Correlation: ', compute_correlation_coefficient(x, y))
```

```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer

vi_data_df = pd.read_csv('vi_text_retrieval.csv')

context = vi_data_df['text']
context = [doc.lower() for doc in context]

tfidf_vectorizer = TfidfVectorizer()
context_embedded = tfidf_vectorizer.fit_transform(context)

print(context_embedded.toarray()[7][0])


```

```python
# Get vocabs
vocabs = tfidf_vectorizer.vocabulary_
# display(vocabs)
# print(len(vocabs))
sorted(vocabs.items(), key=lambda k: k[1])
```

```python
# Question 11:
def tfidf_search(question, tfidf_vectorizer, top_d=5):
    # lowercasing before encoding
    query_embedded = tfidf_vectorizer.transform([question.lower()])
    cosine_scores = cosine_similarity(context_embedded, query_embedded)

    # Get top k cosine score and its index
    results = []
    for idx in cosine_scores.argsort()[-top_d:][::-1]:
        doc_score = {
            'id': idx,
            'cosine_score': cosine_scores[idx]
        }
        results.append(doc_score)

    return results


question = vi_data_df.iloc[0]['question']
print(question)
results = tfidf_search(question, tfidf_vectorizer, top_d=5)
print(results[0]['cosine_score'])
```

```python
# Question 12:
def corr_search(question, tfidf_vectorizer, top_d=5):
    # lowercaseing before encodidng
    query_embedded = tfidf_vectorizer.transform([question.lower()])
    corr_scores = np.corrcoef(
        query_embedded.toarray(),
        context_embedded.toarray()
    )
    corr_scores = corr_scores[0][1:]

    # Get top k correlation score and its indes
    results = []
    for idx in corr_scores.argsort()[-top_d:][::-1]:
        doc = {
            'id': idx,
            'corr_score': corr_scores[idx]
        }
        
        results.append(doc)

    return results

question = vi_data_df.iloc[0]['question']
results = corr_search(question, tfidf_vectorizer, top_d=5)
results[1]['corr_score']
```
